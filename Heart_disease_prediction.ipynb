Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the csv file to dataframe
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
#Print the first 5 rows of the dataset
heart_data.head()
```

```python
#Print the last 5 rows of our dataset
heart_data.tail()
```

```python
# Number of rows and columns in our dataset
heart_data.shape
```

```python
# Getting some info about the data
heart_data.info()
```

```python
#Checking for missing values
heart_data.isna().sum()
```

```python
# Statistical measures about the data
heart_data.describe()
```

```python
# Checking the distrubution of the target variable
heart_data['target'].value_counts()
```

1 --> defective heart

0 --> Healthy heart

Spliting the features and target

```python
x = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

Standardize the data


```python
# Import the standardscaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

Spliting the training data and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train)
print(x_test)
```

Model training

```python
model = LogisticRegression()
```

```python
#Training the Logistic Regression model with the Training data
model.fit(x_train, y_train)
```

Model Evaluation

Accuracy score

```python
x_train_predict = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_predict, y_train)
```

```python
print('Accuracy score of the traing data: ', x_train_accuracy)
```

```python
x_test_predict = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predict, y_test)
```

```python
print('Accuracy score of testing data: ', x_test_accuracy)
```

Building the prediction system

```python
input_data = (63,1,3,145,233,1,0,150,0,2.3,0,0,1)

#Change the data as a np array
input_data_as_np_array = np.asarray(input_data)

#Reshape the np array as we are predicting for only one instance
input_data_reshaped = input_data_as_np_array.reshape(1,-1)

#Standardized the data
input_data_reshaped = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_reshaped)
print(prediction)
```

```python
if (prediction[0] == 0):
  print('The person does not have a heart disease')
else:
  print('The person has a heart disease')
```
