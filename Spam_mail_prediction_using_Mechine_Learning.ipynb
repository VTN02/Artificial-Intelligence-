```python
# Import the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collectioon and Preprocessing

```python
# Loading the data from csv file to a panda dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
raw_mail_data.head()
```

```python
# Checking the no of rows and columns in the pandas dataFrame

raw_mail_data.shape
```

```python
# Label the spam mail as 0, ham mail as 1
raw_mail_data.loc[raw_mail_data['Category'] == 'spam', 'Category'] = 0
raw_mail_data.loc[raw_mail_data['Category'] == 'ham', 'Category'] = 1
```

Spam --> 0

Ham --> 1

```python
raw_mail_data.head()
```

```python
# Saparating the data as text and labels

x = raw_mail_data['Message']
y = raw_mail_data['Category']
```

```python
print(x)
```

```python
print(y)
```

Spliting the data into training data ad testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
# Transform the data to feature vectors that can be used as input to the Logistic Regression

feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)
```

```python
# Convert the y train and y test as integers

y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(y_train)
```

```python
print(x_train_features)
```

Training the model - Using Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

Evaluating the training model

```python
predicting_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train, predicting_on_training_data)
```

```python
print("Accuracy on training data: ", accuracy_on_training_data)
```

```python
predicting_on_test_data = model.predict(x_test_features)
accuracy_on_test_data = accuracy_score(y_test, predicting_on_test_data)
```

```python
print("Accuracy on testing data: ", accuracy_on_test_data)
```

Building the predicting system

```python
input_main = ["Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got amore wat..."]
# Convert text to features
input_feature_data = feature_extraction.transform(input_main)

# Making prediction
prediction = model.predict(input_feature_data)

print(prediction)

if(prediction[0] == 0):
  print('Ham mail')
else:
  print('Spam mail')
```
