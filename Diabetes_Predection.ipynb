```python
# import the dependencies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset.isna().sum()
```

```python
diabetes_dataset.value_counts(['Outcome'])
```

```python
x = diabetes_dataset.drop(columns='Outcome', axis=1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
diabetes_dataset.describe()
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
diabetes_dataset_standardized = scaler.transform(x)
```

```python
diabetes_dataset_standardized
```

```python
x = diabetes_dataset_standardized
y = diabetes_dataset['Outcome']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

```python
x_train_predection = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_predection, y_train)
```

```python
print("Accuracy score for the training data: ", x_train_accuracy)
```

```python
x_test_predection = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predection, y_test)
```

```python
print("Accuracy score of testing data: ", x_test_accuracy)
```

```python
# Make prediction system for the model
input_data = (1,85,66, 29,0,26.6,	0.351,	31)

input_data_as_np_array = np.asarray(input_data)

input_data_reshaped = input_data_as_np_array.reshape(1, -1)

std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)

print(prediction)

if(prediction[0] == 0):
  print("The person is not diabetic")
else:
  print("The person is diabetic")
```
