Resizing the Image using pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpeg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# Displaying the image from nparray

img_res = npimg.imread('/content/dog_image_resized.jpg')
img_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting RGB images to GrayScale image using OpenCV

```python
# Importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/images.jpeg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Displayin the image
cv2_imshow(grayscale_image)
```

```python
# Saving the grayscale image
cv2.imwrite('dog_gray_image.jpg', grayscale_image)
```
